```python
import time
import gym
import numpy as np
```

## Train a taxi

```python
env = gym.make('Taxi-v3', render_mode = "human")
```

```python
alpha = 0.5
gamma = 0.99
epsilon = 0.01
max_epsilon = 1.0
min_epsilon = 0.01
decay_rate = 0.9
counter = 0
```

```python
num_states = env.observation_space.n
num_actions = env.action_space.n
Q_tabel = np.zeros((num_states, num_actions))
Q_tabel
```

```python
num_episodes = 1000
num_steps = 100
for episode in range(num_episodes):
    counter += 1
    state = env.reset()[0]
    total_reward = 0
    for cur_step in range(num_steps):

        # e-greedy choice, pick up action to generate samples

        if np.random.uniform(0,1) < epsilon:

            # every action can be chosen from this line: e * 1/num_actions.
            cur_action = env.action_space.sample()
        else:
            cur_action = np.argmax(Q_tabel[state, : ])
        # generate complete sample by current q_table. choose action with largest q_value or e-greedy.
        new_state, reward, isdone, _, info = env.step(cur_action)
        total_reward += reward
        # update q_table
        Q_tabel[state, cur_action] += alpha * (reward + gamma * np.max(Q_tabel[new_state, : ]) - Q_tabel[state, cur_action])

        state = new_state
        if isdone:
            break

    epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate * episode)
    print("current episode", episode, "current reward", total_reward)
np.save(file="Q_table_taxi.npy", arr=Q_tabel)
print("trainning complete")
```

```python

cur_Q_tabel = np.load(file="Q_table_taxi.npy")

for episode in range(50):
    env = gym.make('Taxi-v3', render_mode="human")
    state = env.reset()[0]
    done = False
    reward = 0
    
    for cur_step in range(num_steps):

        print("current step is {}".format(cur_step+1))

        cur_action = np.argmax(cur_Q_tabel[state, : ])
        new_state, reward, isdone, _, info = env.step(cur_action)
        reward += reward

        print("current score: {}".format(reward))
        state = new_state
        
        if isdone:
            break
```
